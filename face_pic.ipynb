```python
import cv2 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/home/navya/Documents/to_study/face_detection/data/mammuty.jpeg')
```

```python
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

```python
cv2.imshow('img',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img_grey = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
face_cascade = cv2.CascadeClassifier('/home/navya/Documents/to_study/face_detection/data/haarcascade_frontalface_default.xml')
```

```python
faces = face_cascade.detectMultiScale(img_grey)
```

```python
#face count
len(faces)
```

```python
faces #these values are points x,y,w,h
```

```python
for (x,y,w,h) in faces:   #width hight eplum add akanam
    cv2.rectangle(img,(x,y),(x+w,y+h),color=(0,255,0),thickness=1)
```

```python
cv2.imshow('face detect',img)
cv2.waitKey()
cv2.destroyAllWindows()
```
